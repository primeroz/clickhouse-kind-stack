```python
import s3fs
import pyarrow.parquet as pq
from sqlalchemy import create_engine



fs = s3fs.S3FileSystem(anon=True)

bucket = 'aws-public-blockchain'
path = 'v1.0/btc/transactions/date=2023-08-10/802480.snappy.parquet' #if its a directory omit the traling /
bucket_uri = f's3://{bucket}/{path}'

dataset = pq.ParquetDataset(bucket_uri, filesystem=fs)
table = dataset.read()
df = table.to_pandas() 
```
