```python
import s3fs
import pyarrow.parquet as pq
import pandas as pd


fs = s3fs.S3FileSystem(anon=True)

bucket = 'aws-public-blockchain'
path = 'v1.0/btc/transactions/date=2023-08-10/802480.snappy.parquet' #if its a directory omit the traling /
bucket_uri = f's3://{bucket}/{path}'

dataset = pq.ParquetDataset(bucket_uri, filesystem=fs)
table = dataset.read()
df = table.to_pandas() 
```

```python
from pprint import pprint
import numpy

def normalize(row):
    # Do i need this copy or can i just work on row ?
    _row = row.copy(deep=True)

    _inputs = row.get('inputs')
    ## Explode Imputs    
    inputs = {}
    
    if row.get('is_coinbase'):
        ## TODO need to actually add the expected columns anyway
        pass
    else:
        for _i in _inputs:
            for _k, _v in _i.items():
                inputs.setdefault("inputs."+_k, []).append(_v.tolist() if isinstance(_v, numpy.ndarray) else _v)
            

    #pprint(inputs)
    ## Explode Outputs
    _outputs = row.get('outputs')

    outputs = {}
    for _o in _outputs:
        for _k, _v in _o.items():
            outputs.setdefault("outputs."+_k, []).append(_v.tolist() if isinstance(_v, numpy.ndarray) else _v)

    #pprint(outputs)
    ## drop original columns
    _row.drop(['inputs', 'outputs'], inplace=True)

    # print(type(_row))
    # print(_row)
    # print(type(pd.Series(inputs)))
    # print(pd.Series(inputs))
    
    return pd.concat([_row, pd.Series(inputs), pd.Series(outputs)])

normalized_items = items.apply(normalize, axis=1)

normalized_items
```

```python
#normalized_items.to_csv("output.csv", index=False, quotechar='"',doublequote=False, na_rep="None")
normalized_items.to_parquet("output.parquet", index=None)
```
